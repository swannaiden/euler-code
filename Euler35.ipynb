```python
# generate all the prime numbers below 1 million
import math
primes = [2]

for i in range(3,10**6, 2):
  isPrime = True
  for j in range(2, math.floor(math.sqrt(i)+1)):
    if(i%j == 0):
      isPrime = False
      break
  if(isPrime):
    primes.append(i)

print(primes[1:100])
```

```python
# this is for numbers which are three digits

hash = []
multl = []
addl = []
digits = []

for prime in primes2:
  digit = prime
  add = 0
  mult = 1
  num_digits = 0
  while(digit > 0):
    add = add + digit % 10
    mult = mult * (digit % 10)
    digit = math.floor(digit/10)
    num_digits = num_digits +1 

  hash.append((add, mult))
  multl.append(mult)
  addl.append(add)
  digits.append(num_digits)
```

```python
zip1 = zip(addl, primes2)
zip1_sort = sorted(zip1)
# zip1_sort

zip2 = zip(multl, primes2)
zip2_sort = sorted(zip2)
# print(zip2_sort)

zip3 = zip(digits, addl, multl, primes2)
zip3_sort = sorted(zip3, key = lambda x: (x[0], x[1], x[2]))
```

```python
valids = []

for i in range(0, len(zip3_sort)):
  # first check to make sure that there are enough with that specfic combination
  valid = True
  for j in range(0, zip3_sort[i][0]):
    if(i+j >= len(zip3_sort)):
      valid = False
      break
    if(zip3_sort[i+j][1]!=zip3_sort[i][1] or zip3_sort[i+j][2]!=zip3_sort[i][2]):
      valid = False
      break
  if(valid == True):
    for j in range(0, zip3_sort[i][0]):
      valids.append(zip3_sort[i+j])
  
```
